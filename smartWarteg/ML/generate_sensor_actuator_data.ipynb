```python
import csv
import random

def generate_dish_data():
    with open('dish.csv', mode='w', newline='') as csvfile:
        fieldnames = ['dish_level_detector', 'waterflow_monitor', 'dish_weight_reader', 
                      'dish_cleaner_rotator', 'sink_water_valve', 'dish_soap_pump']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()

        for _ in range(1000):
            dish_level_detector = random.randint(0, 10)
            waterflow_monitor = random.randint(0, 10)
            dish_weight_reader = random.randint(0, 10)

            dish_cleaner_rotator = (dish_level_detector + dish_weight_reader)//2
            sink_water_valve = -1 if (dish_level_detector > 7 or dish_weight_reader > 8) else -2
            dish_soap_pump = (dish_level_detector + dish_weight_reader)//4

            writer.writerow({
                'dish_level_detector': dish_level_detector,
                'waterflow_monitor': waterflow_monitor,
                'dish_weight_reader': dish_weight_reader,
                'dish_cleaner_rotator': dish_cleaner_rotator,
                'sink_water_valve': sink_water_valve,
                'dish_soap_pump': dish_soap_pump
            })

if __name__ == "__main__":
    generate_dish_data()
```

```python
import csv
import random

def generate_stove_data():
    with open('stove.csv', mode='w', newline='') as csvfile:
        fieldnames = ['temperature_reader', 'gas_leak_detector', 'smoke_detector', 
                      'water_sprinkler', 'fire_alarm', 'gas_exhaust']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()

        for _ in range(1000):
            temperature_reader = random.randint(20, 120)
            gas_leak_detector = random.randint(0, 10)
            smoke_detector = random.randint(0, 10)

            water_sprinkler = (temperature_reader - 50) + smoke_detector * 3 if (temperature_reader >= 50) else 0
            fire_alarm = -1 if(temperature_reader >= 50 and gas_leak_detector >= 0 and smoke_detector > 7) else -2
            gas_exhaust =  gas_leak_detector*5+smoke_detector*5 

            writer.writerow({
                'temperature_reader': temperature_reader,
                'gas_leak_detector':  gas_leak_detector,
                'smoke_detector': smoke_detector,
                'water_sprinkler': water_sprinkler,
                'fire_alarm': fire_alarm,
                'gas_exhaust': gas_exhaust
            })

if __name__ == "__main__":
    generate_stove_data()
```

```python
import csv
import random

def generate_customer_data():
    with open('customer.csv', mode='w', newline='') as csvfile:
        fieldnames = ['camera', 'motion_detector', 'sound_sensor', 
                      'sound_alert', 'light', 'monitor']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()

        for _ in range(1000):
            camera = random.randint(0, 10)
            motion_detector = random.randint(0, 10)
            sound_sensor = random.randint(1, 100 )

            sound_alert = -2
            if motion_detector > 5 and sound_sensor > 50:
                sound_alert = -1
            light = -1 if (camera > 0) else -2
            monitor = camera

            writer.writerow({
                'camera': camera,
                'motion_detector':  motion_detector,
                'sound_sensor': sound_sensor,
                'sound_alert': sound_alert,
                'light': light,
                'monitor': monitor
            })

if __name__ == "__main__":
    generate_customer_data()
```
